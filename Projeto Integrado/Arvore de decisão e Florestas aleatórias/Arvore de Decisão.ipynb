```python
# Implementação de uma árvore de decisão
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets

iris = datasets.load_iris()
features = iris.data
target = iris.target

decision_tree = DecisionTreeClassifier()

model = decision_tree.fit(features, target)
```

A árvore de decisão é um modelo de classificação que procura por regras de decisão que produz a maior redução de impuridade dos dados

A entropia é medida pelo indice de Gini, que pode ser calculado como:
   $$G(t) = 1 - \sum_{i=1}^{c} p_i²$$
Onde $t$ é um nó e $p_i$ é a proporção de observações da classe $c$ no nó $t$

Também é possivel escolher a entropia como métrica de critéria do modelo
passando  `DecisionTreeClassifier(criterion='entropy')`

```python
# Vizualizando o modelo
from IPython.display import Image
from sklearn import tree
import pydotplus

dot_data = tree.export_graphviz(
    decision_tree,
    out_file=None,
    feature_names=iris.feature_names,
    class_names=iris.target_names
)

graph = pydotplus.graph_from_dot_data(dot_data)

Image(graph.create_png())
```

```python
graph.write_png("exemplo_arvore.png")
```
